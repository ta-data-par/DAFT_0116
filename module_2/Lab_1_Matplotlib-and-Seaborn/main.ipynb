#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# Create the data arrays
x = np.linspace(0, 5, 100)  # Using linspace for a smoother curve
y1 = x**2
y2 = np.exp(x)

# Create a figure and two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))  # Adjust figsize as needed

# Plot data on the first subplot (normal scale)
axes[0].plot(x, y1, label='y = x^2')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Sculder')
axes[0].legend()

# Plot data on the second subplot (logarithmic scale)
axes[1].plot(x, y2, label='y = e^x')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Scally')
axes[1].set_yscale('log')  # Set logarithmic scale for the y-axis
axes[1].legend()

# Add space between subplots to avoid overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# Convert the date column to a datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Extract the month from the datetime column
df['Month'] = df['Date'].dt.month

# Calculate the median steps for each month
median_steps_by_month = df.groupby('Month')['Steps'].median()

# Create a plot to visualize the median steps by month
plt.figure(figsize=(9, 5))
plt.bar(median_steps_by_month.index, median_steps_by_month.values)
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.title('Median Steps by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

# Create a plot to show the average steps on weekends vs. workdays
df['Day_of_Week'] = df['Date'].dt.dayofweek
df['Is_Weekend'] = df['Day_of_Week'].isin([5, 6])  # Saturday = 5, Sunday = 6

steps_weekend = df[df['Is_Weekend']]['Steps'].mean()
steps_workdays = df[~df['Is_Weekend']]['Steps'].mean()

plt.figure(figsize=(7, 4))
plt.bar(['Weekend', 'Workdays'], [steps_weekend, steps_workdays])
plt.xlabel('Day Type')
plt.ylabel('Average Steps')
plt.title('Average Steps on Weekends vs. Workdays')
plt.show()

```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# Create a list of feature pairs
feature_pairs = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

#  Create a loop to plot the scatter plots
plt.figure(figsize=(10, 5))
for i, (x_feature, y_feature) in enumerate(feature_pairs, 1):
    plt.subplot(2, 2, i)
    plt.scatter(df[x_feature], df[y_feature], alpha=0.5)
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.title(f'{x_feature} vs. {y_feature}')

plt.tight_layout()
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# Set the plot style to 'classic'
plt.style.use('classic')

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Fare'], df['Age'], color='deepskyblue', alpha=0.5)
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Scatter Plot: Fare vs. Age')
plt.show()

# Joint Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Fare'], df['Age'], color='deepskyblue', alpha=0.5)
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Joint') 
```

```python
# Method 2 - seaborn
# Set the plot style to 'whitegrid'
sns.set_style('whitegrid')

# Method 1: Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Fare', y='Age', data=df, color='deepskyblue', alpha=0.5)
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Scatter Plot: Fare vs. Age')
plt.show()

# Method 2: Joint Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Fare', y='Age', data=df, color='deepskyblue', alpha=0.5)
sns.histplot(df['Fare'], bins=20, color='gray', alpha=0.6, kde=True)
sns.histplot(df['Age'], bins=20, color='orange', alpha=0.6, kde=True)

plt.legend(['Scatter', 'Fare KDE', 'Age KDE'])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Joint Plot: Fare vs. Age')
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
# Drop non-numeric columns from the DataFrame
numeric_df = df.select_dtypes(include='number')

# Calculate the correlation matrix for the numeric columns
correlation_matrix = numeric_df.corr()

# Set the plot style to 'whitegrid'
sns.set_style('whitegrid')

# Create a heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
